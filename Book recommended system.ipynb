Pivot table

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='Title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

now we will cluster it using algorithm but first we will convert this matrix into sparse matrix.
in book_pivot we have lot of zeros thats why we will do spasre. this will only consider non zero values

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')   #brute will calculate distance from every point
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1,),n_neighbors=6)
```

Now lastly we will create an function in whch we will give our book name and it will recommend us 5 more books.

```python
book_pivot.index[20]
```

```python
np.where(book_pivot.index=='A Painted House')[0][0]
```

```python
def recommend_books(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1,),n_neighbors=6)
    for i in range(len(suggestions)):
        print(f'The Suggestions for {book_name} are: ')
        books = book_pivot.index[suggestions[i]]
        
    return books[1:]
    
```

```python
recommend_books('Animal Farm')
```
